```python
import pandas as pd
from bs4 import BeautifulSoup
import numpy as np
import re
import openpyxl
from config.want_type import *
from PyQt5.QtWidgets import *# 메인 루프
import os
import sys
from selenium import webdriver
```

```python
class CherryPick():
    def __init__(self):
        self.krx_code_list = []
        self.financial_dict = {}
        self.get_code()
        self.stock_info()
        #### 외부클래스/변수 ###
        want_code = Want_code()
        self.want_area_list = want_code.want_area #원하는 업종리스트
    def get_code(self):
        self.df = pd.read_excel("KrxCode.xlsx")
        self.krx_code_list = self.df["단축코드"].values
    def stock_info(self): # 개별 종목코드의 재무지표를 얻는다
        idx = 0
        for code in self.krx_code_list:
            idx+=1
            print("{}/850 진행중".format(idx,len(self.krx_code_list)))
            self.crawl_info(code)
    def crawl_info(self,code): # 개별 종목코드의 재무지표를 얻기위해 크롤링을 수행하는 메소드
        try:
            chrome_driver = "C:/Users/user/Downloads/chromedriver_win32/chromedriver.exe"
            url = "https://finance.naver.com/item/main.nhn?code=" + code
            options = webdriver.ChromeOptions()
            options.add_argument("headless") #크롬창이 안켜지도록
            options.add_argument("start-maximized")
            options.add_argument("disable-gpu")
            options.add_argument("disable-infobars")
            options.add_argument("--disable-extensions")
            # 속도 향상을 위한 옵션 해제
            prefs = {'profile.default_content_setting_values': {'cookies': 2, 'images': 2, 'plugins': 2, 'popups': 2,
                                                                'geolocation': 2, 'notifications': 2,
                                                                'auto_select_certificate': 2, 'fullscreen': 2,
                                                                'mouselock': 2, 'mixed_script': 2, 'media_stream': 2,
                                                                'media_stream_mic': 2, 'media_stream_camera': 2,
                                                                'protocol_handlers': 2, 'ppapi_broker': 2,
                                                                'automatic_downloads': 2, 'midi_sysex': 2,
                                                                'push_messaging': 2, 'ssl_cert_decisions': 2,
                                                                'metro_switch_to_desktop': 2,
                                                                'protected_media_identifier': 2, 'app_banner': 2,
                                                                'site_engagement': 2, 'durable_storage': 2}}
            options.add_experimental_option('prefs', prefs)
            driver = webdriver.Chrome(chrome_driver,options = options)
            driver.get(url)
            src = driver.page_source
            driver.quit()
            soup = BeautifulSoup(src,features="lxml")
            code_nm = self.df[self.df["단축코드"] == code]["한글 종목명"].values[0]
            area = soup.select("div > h4 > em > a")[0].text
            area_per = float(soup.select("tr.strong > td > em")[2].text)
            cap = soup.select('table[class = "tb_type1 tb_num"]')[0].select("tbody > tr")[3].find("td").text
            if "," in cap:
                cap = int(re.sub(",", "", cap))
            else:
                cap = int(cap)
            per = soup.select('table[class = "tb_type1 tb_num"]')[0].select("tbody > tr")[12].find("td").text
            if per is None:
                per = 0
            else:
                per = float(per)
            pbr = soup.select('table[class = "tb_type1 tb_num"]')[0].select("tbody > tr")[13].find("td").text
            if pbr is None:
                pbr = 0
            else:
                pbr = float(pbr)
            tr = soup.find("table", {"class": "tb_type1 tb_num tb_type1_ifrs"}).select("tbody > tr")
            if tr is not None:
                total_list = []
                for i in np.array([0,1,2,3,5,6,7]):
                    td = tr[i].findAll("td")
                    data= td[8].text.strip()
                    if "," in data:
                        data = float(re.sub(",", "", data))
                        total_list.append(data)
                    elif data == None:
                        total_list.append(0)
                    else:
                        total_list.append(float(data))
                if code not in self.financial_dict:
                    self.financial_dict.update({code:{}})
                self.financial_dict[code].update({
                                                 "종목명" : code_nm,
                                                  "시가총액": cap,
                                                 "업종" : area,
                                                 "업종PER" : area_per,
                                                 "매출액" : total_list[0],
                                                  "영업이익" : total_list[1],
                                           
                                          "당기손익" : total_list[2],
                                                  "영업이익률" : total_list[3],
                                                  "ROE" : total_list[4],
                                                  "부채비율" : total_list[5],
                                                  "당좌비율" : total_list[6],
                                                  "PER" : per,
                                                  "PBR" : pbr
                })
        except :
            pass
    def to_xlsx(self):
        if not os.path.exists("CherryPick_KR_S.xlsx"):
            wb = openpyxl.Workbook()
            sheet1 = wb.active
            sheet2 = wb.create_sheet()
            sheet3 = wb.create_sheet()
            sheet1.title = "전종목"
            sheet2.title = "선택업종"
            sheet3.title = "저PER고ROE"
            sheet1.append(["종목코드", "종목명", "시가총액", "PER", "PBR",
                           "업종", "업종PER", "매출액", "영업이익", "당기손익",
                           "영업이익률", "ROE", "당좌비율", "부채비율", "기준_최근분기"])
            sheet2.append(["종목코드", "종목명", "시가총액", "PER", "PBR",
                           "업종", "업종PER", "매출액", "영업이익", "당기손익",
                           "영업이익률", "ROE", "당좌비율", "부채비율", "기준_최근분기"])
            sheet3.append(["종목코드", "종목명", "시가총액", "PER", "PBR",
                           "업종", "업종PER", "매출액", "영업이익", "당기손익",
                           "영업이익률", "ROE", "당좌비율", "부채비율", "기준_최근분기"])
        else :
            wb = openpyxl.load_workbook("CherryPick_KR.xlsx")
            sheet1 = wb.active
            sheet2 = wb["선택업종"]
            sheet3 = wb["저PER고ROE"]
        idx = 0
        for code in self.financial_dict.keys():
            idx += 1
            print("{}/{} 진행중".format(idx, len(self.financial_dict)))
            code_nm = self.financial_dict[code]["종목명"]
            cap = self.financial_dict[code]["시가총액"]
            per = self.financial_dict[code]["PER"]
            pbr = self.financial_dict[code]["PBR"]
            area = self.financial_dict[code]["업종"]
            areaper = self.financial_dict[code]["업종PER"]
            sales = self.financial_dict[code]["매출액"]
            profit = self.financial_dict[code]["영업이익"]
            earning = self.financial_dict[code]["당기손익"]
            sales_ratio= self.financial_dict[code]["영업이익률"]
            roe = self.financial_dict[code]["ROE"]
            dang = self.financial_dict[code]["당좌비율"]
            debt = self.financial_dict[code]["부채비율"]
            sheet1.append([code,code_nm,cap,per,pbr,area,areaper,sales,profit,earning,sales_ratio,roe,dang,debt])
            if area in self.want_area_list:
                sheet2.append(
                    [code, code_nm, cap, per, pbr, area, areaper, sales, profit, earning, sales_ratio, roe, dang, debt])
            if per <=20 and roe >=15:
                sheet3.append(
                    [code, code_nm, cap, per, pbr, area, areaper, sales, profit, earning, sales_ratio, roe, dang, debt])
        wb.save("CherryPick_KR_S.xlsx")
        print("완료")
        sys.exit()
if __name__ == "__main__":
    app = QApplication(sys.argv)
    cherry = CherryPick()
    cherry.to_xlsx()
    app.exec_()
[출처] 파이썬과 금융 - 네이버 금융 재무지표 크롤링_selenium 사용      (코스피 코스닥 종목)|작성자 임자밀
```
